# outliers

```python
#finding outliers and filling them
col=['Current Loan Amount','Annual Income','Monthly Debt','Current Credit Balance','Number of Open Accounts']
for i in col:
    q1=credit_df[i].quantile(0.25)
    q3=credit_df[i].quantile(0.75)
    Iqr=q3-q1
    upper=q3+1.5*Iqr
    lower=q1-1.5*Iqr
    credit_df[i]=credit_df[i].apply(lambda x:lower if x<lower else (upper if x>upper else x))
```

```python
credit_df.fillna({'Maximum Open Credit':credit_df['Maximum Open Credit'].mean()},inplace=True)
```

```python
credit_df['Years in current job']=credit_df['Years in current job'].apply(lambda x:0.5 if x=='< 1 year' else (11 if x=='10+ years' else (x[:1] if pd.isna(x)==False else x)))
credit_df['Years in current job']=pd.to_numeric(credit_df['Years in current job'],downcast='integer')
credit_df.fillna({'Years in current job':credit_df['Years in current job'].mean()},inplace=True)
```

```python
credit_df['Annual Income']=credit_df['Annual Income'].fillna(credit_df['Annual Income'].mean())
credit_df.fillna({'Bankruptcies':credit_df['Bankruptcies'].mode()[0]},inplace=True)
credit_df.fillna({'Tax Liens':credit_df['Tax Liens'].mode()[0]},inplace=True)
```

```python
credit_df['Credit Score']=credit_df['Credit Score'].apply(lambda x:np.nan if x>900 else x)
```

# KNN

```python
x=credit_df[['Current Loan Amount', 'Term',
       'Credit Score', 'Annual Income', 'Years in current job',
       'Home Ownership', 'Purpose', 'Monthly Debt', 'Years of Credit History',
       'Number of Open Accounts', 'Number of Credit Problems',
       'Current Credit Balance', 'Maximum Open Credit', 'Bankruptcies',
       'Tax Liens' ]]
y=credit_df['Loan Status']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import LinearSVC
from sklearn.metrics import confusion_matrix,mean_squared_error,r2_score,classification_report,accuracy_score
```

```python
x_num_cols=x.select_dtypes(include='number')
x_cat_cols=x.select_dtypes(include='object')
```

```python
x_cat_cols=pd.get_dummies(data=x_cat_cols)
```

```python
x=pd.concat([x_cat_cols,x_num_cols],axis=1)
```

```python
x
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=7)
```

```python
from sklearn.model_selection import RandomizedSearchCV
import numpy as np

# Define parameter distribution for Randomized Search
param_dist = {
    'n_neighbors':np.arange(5,1000)
}

# Initialize RandomizedSearchCV
random_search_svm = RandomizedSearchCV(
    estimator=KNeighborsClassifier(),
    param_distributions=param_dist,
    n_iter=50,  # Number of parameter settings sampled
    scoring='accuracy',  # Evaluation metric
    cv=5,  # 5-fold cross-validation
    random_state=42,
    verbose=2,
    n_jobs=-1  # Use all available cores
)

# Fit the model on training data
random_search_svm.fit(x_train, y_train)

# Print the best parameters
print("Best parameters found (Randomized Search):", random_search_svm.best_params_)
```
